```python
from tqdm import tqdm
import glob
import os.path

intersection = False

for annotation_path in glob.glob("resources/data/**/*.ann", recursive=True):
    text_path = os.path.join(os.path.dirname(annotation_path), os.path.basename(annotation_path).split(".")[0] + ".txt")
    with open(text_path, "r") as text_file:
        symbol_count = len(text_file.read())

    visited = [0] * symbol_count

    with open(annotation_path, "r") as annotation_file:
        annotation_lines = annotation_file.readlines()
        for annotation_line in tqdm(annotation_lines):
            annotation_data = annotation_line.split()
            if not annotation_data[0].startswith("T"):
                continue
            start = int(annotation_data[2])
            end = int(annotation_data[3])
            for i in range(start, end):
                if visited[i] == 1:
                    print("Intersection!: ", annotation_line)
                    intersection = True
                visited[i] = 1

if not intersection:
    print("There is no intersection!")
```
